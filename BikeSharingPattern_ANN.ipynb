# **Predicting Bike Sharing Patterns Using ANN**

```python
# https://www.kaggle.com/datasets/itssuru/bike-sharing-system-washington-dc

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
```

```python
from google.colab import drive
drive.mount("/content/gdrive")
```

```python
df_path = '/content/gdrive/MyDrive/Deep Learning Datasets/Bike Sharing/train_bikes.csv'
df = pd.read_csv(df_path)
df.head()
```

CONVERSION OF STRING DATA SET

```python
df["datetime"] = df["datetime"].astype("datetime64")
```

```python
df["hour"] = df["datetime"].dt.hour
df["day"] = df["datetime"].dt.day
df["month"] = df["datetime"].dt.month
df["year"] = df["datetime"].dt.year
```

```python
df.head()
```

```python
df = df.drop(["datetime"], axis=1)
```

```python
y = df["count"]
X = df.drop(["count"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=64)
```

NORMALIZING THE DATA SET

```python
data = (X_train - X_train.mean()) / X_train.std()
y.head()
```

```python
X_train
```

A Sequential model is appropriate for a plain stack of layers where each layer has exactly one input tensor and one output tensor.

```python
model = Sequential()
model.add(Dense(128, activation="relu"))
model.add(Dense(20, activation="linear"))
model.add(BatchNormalization())
model.add(Dense(16, activation="relu"))
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")  #gradient descent
```

```python
model.fit(X_train,y_train, validation_split=0.2, epochs = 300, verbose = 1)
```

```python
model.evaluate(X,y)
```

```python
X_test
```

```python
test = (X_test - X_test.mean())/X_test.std()
print(test)
```

```python
test = np.asarray(test).astype(np.float32)
```

```python
y_pred = model.predict(test)
```

```python
model.evaluate(X_test, y_test)
```

```python
r2 = r2_score(y_test, y_pred)
print(r2)
```
